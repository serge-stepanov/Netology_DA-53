Ответьте на следующие вопросы:

Как критики относятся к спортивным играм?

Критикам нравятся больше игры на PC или на PS4?

Критикам больше нравятся стрелялки или стратегии?

Для каждого вопроса:

сформулируйте нулевую и альтернативную гипотезы; выберите пороговый уровень статистической значимости; опишите полученные результаты статистического теста.

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/vgsales.csv')
df.head()
```

```python
df.info()
```

Среднее по всем играм - 70, так что ставим условие выше этого среднего

H0: средняя удовлетворенность критиками спортивными играми <= 70

H1: средняя удовлетворенность критиками спортивными играми более 70

```python
df_sport = df.loc[(df['Genre'] == 'Sports')&(df['Critic_Score'] > 0)]
df_sport.head()
```

```python
from scipy import stats as st
alpha = 0.05
result = st.ttest_1samp(df_sport['Critic_Score'], 70, alternative='greater')

print(df_sport['Critic_Score'].mean())
print(result)

if result.pvalue < alpha: 
    print('Отвергаем нулевую гипотезу, среднее больше 70 (игры нравятся)')
else:
    print('Не отвергаем нулевую гипотезу')
```

Получается, что в среднем критики ставят высокие оценки спортивным играм

H0: оценки критиков игры на PC не отличаются от оценок игр на PS4

H1: оценки критиков игры на PC отличаются от оценок игр на PS4

```python
df_pc = df.loc[(df['Platform'] == 'PC')&(df['Critic_Score'] > 0)]
df_pc.head()
```

```python
df_ps = df.loc[(df['Platform'] == 'PS4')&(df['Critic_Score'] > 0)]
df_ps.head()
```

```python
alpha = 0.05
result = st.ttest_ind(df_pc['Critic_Score'], df_ps['Critic_Score'], equal_var=False)

print('PC', df_pc['Critic_Score'].mean())
print('PS4', df_ps['Critic_Score'].mean())

print(result)

if result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу, критики по разному относятся к платформам')
else:
    print('Не отвергаем нулевую гипотезу, у критиков есть предпочтения между двумя платформами')
```

```python
if (df_pc['Critic_Score'].mean() < df_ps['Critic_Score'].mean()):
    print('Критикам больше нравятся игры на PS4')
else:
    print('Критикам больше нравятся игры на PC')
```

pvalue меньше 0.05 - различия не существенны

H0: оценки критиков для стрелялок не отличаются от оценок стратегий

H1: оценки критиков для стрелялок отличаются от оценок стратегий

```python
df_action = df.loc[(df['Genre'] == 'Shooter')&(df['Critic_Score'] > 0)]
df_action.head()
```

```python
df_strategy = df.loc[(df['Genre'] == 'Strategy')&(df['Critic_Score'] > 0)]
df_strategy.head()
```

```python
alpha = 0.05
result = st.ttest_ind(df_action['Critic_Score'], df_strategy['Critic_Score'], equal_var=False)

print('Shooter', df_action['Critic_Score'].mean())
print('Strategy', df_strategy['Critic_Score'].mean())

print(result)

if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, критики по разному относятся к выбранным жанрам')
else:
    print('Не отвергаем нулевую гипотезу, у критиков есть предпочтения между двумя жанрами')

#pvalue меньше 0.05 - различия не существенны
```

```python
if (df_action['Critic_Score'].mean() < df_strategy['Critic_Score'].mean()):
    print('Критикам больше нравятся стратегии')
else:
    print('Критикам больше нравятся стрелялки')
```

```python
df = pd.read_csv('/content/drive/MyDrive/spam.csv')
df.head()
```

```python
df['Message'] = df['Message'].str.lower()
df.head()
```

```python
import re
df['Message'] = df['Message'].apply(lambda x: re.sub('[\W_]+', ' ', x))
df.head()
```

```python
import nltk
from nltk.corpus import stopwords
stopwords_set = set(stopwords.words('english'))
texts_splitted = [text.split() for text in df['Message']]
without_sw = [[word for word in text if word not in stopwords_set] for text in texts_splitted]
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('omw-1.4')
nltk.download('wordnet')

wordnet_lemmatizer = WordNetLemmatizer()
lemmatized_text = [[wordnet_lemmatizer.lemmatize(word) for word in text] for text in without_sw]
```

```python
df['Message'] = [' '.join(row) for row in lemmatized_text]
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df.Message)
names = tfidf.get_feature_names()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=names)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X = tfidf_matrix 
y = df['Category']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_pred) 
# т.е. хорошее сообщение ошибочно 3 раза попало в спам, а спам 66 раз попал в хорошие
```
