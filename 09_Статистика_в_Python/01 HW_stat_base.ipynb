```python
#Необходимо проверить данные на выбросы и пустые ячейкм (NaN)
def ejection(df, param):
    '''
    Проверка датафрейма на наличие выбросов и их очистка
    '''
    q1 = df[param].quantile(0.25)
    q3 = df[param].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    remove_ejection = df[df[param].between(lower_bound, upper_bound, inclusive=True)]
    df_result = pd.concat([df, remove_ejection]).drop_duplicates(keep=False)[[param]]
    df_result = df_result.dropna()
    return df_result[param].tolist()
```

```python
ejection_temp = ejection(df_horse, 'rectal temperature')
ejection_temp
```

```python
ejection_pulse = ejection(df_horse, 'pulse')
ejection_pulse
```

```python
ejection_rate = ejection(df_horse, 'respiratory rate')
ejection_rate 
```
