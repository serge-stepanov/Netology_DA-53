# Задание 1
Дана переменная, в которой хранится слово из латинских букв. Напишите код, который выводит на экран:

среднюю букву, если число букв в слове нечетное;
две средних буквы, если число букв четное.

Примеры работы программы:

word = 'test'

Результат:

es

word = 'testing'

Результат:

t

```python
word = input("Напишите слово: ")
 
length_word = len(word) #определяем длину слова
good_word = length_word / 2 #делим пополам (для удобства вывода нужных букв из середины слова)
if length_word % 2 == 0: #если слово четное, то берем 2 центральные буквы
    print(word[int(good_word)-1]+word[int(good_word)])
elif length_word % 2 !=0: #если слово имеет нечетное кол-во букв, то берем среднюю букву
    print(word[int(good_word)])
```

# Задание 2
Напишите программу, которая последовательно запрашивает у пользователя числа (по одному за раз) и после первого нуля выводит сумму всех ранее введенных чисел.

Примеры работы программы:

Введите число:

1

Введите число:

4

Введите число:

6

Введите число:

0

Результат:

11

Введите число:

0

Результат:

0

```python
number = int(input('Введите число: '))
total = 0

while number !=0: #пока число не равно нулю
    total += number #мы их складываем в переменной Total
    number = int(input('Введите число: '))
print(f'Результат: {total}') #печатаем сумму, когда ввели 0.
```

```python
boys = ['Peter','Alex','John','Arthur','Richard']
girls = ['Kate','Liza','Kira','Emma','Trisha']

if len(boys) != len(girls): #делаем проверку на количество мальчиков и девочек, если не равно, то выводим сообщение
  print('Внимание, кто-то может остаться без пары!')
else:
  boys.sort() #сортируем мальчиков
  girls.sort() #сортируем девочек
  couples = dict(zip(boys, girls)) #создаем новый словарь 
  print('Иеальны пары:')
  for boy, girl in couples.items(): #перебираем циклом соответствующие пары и выводим на печать
       print(f'{boy} и {girl}')

```

# Задание 5 (необязательное)
Дан поток логов по количеству просмотренных страниц для каждого пользователя. Вам необходимо написать алгоритм, который считает среднее значение просмотров на пользователя. Т. е. надо посчитать отношение суммы всех просмотров к количеству уникальных пользователей.

Примеры работы программы:

stream = [

‘2018-01-01,user1,3’,

‘2018-01-07,user1,4’,

‘2018-03-29,user1,1’,

‘2018-04-04,user1,13’,

‘2018-01-05,user2,7’,

‘2018-06-14,user3,4’,

‘2018-07-02,user3,10’,

‘2018-03-21,user4,19’,

‘2018-03-22,user4,4’,

‘2018-04-22,user4,8’,

‘2018-05-03,user4,9’,

‘2018-05-11,user4,11’,

]

Результат:

Среднее количество просмотров на уникального пользователя: 23.25

stream = [

‘2018-01-01,user100,150’,

‘2018-01-07,user99,205’,

‘2018-03-29,user1001,81’

]

Результат:

Среднее количество просмотров на уникального пользователя: 145.33

```python
stream = [
    '2018-01-01,user1,3',
    '2018-01-07,user1,4',
    '2018-03-29,user1,1',
    '2018-04-04,user1,13',
    '2018-01-05,user2,7',
    '2018-06-14,user3,4',
    '2018-07-02,user3,10',
    '2018-03-21,user4,19',
    '2018-03-22,user4,4',
    '2018-04-22,user4,8',
    '2018-05-03,user4,9',
    '2018-05-11,user4,11',
]

views = 0 #создаем переменную для подсчета уникальных просмотров
users = [] #создаем контейнер для пользователей
for user in stream:
    divider = user.split(',') #разделяем запись по уникальному ключу - запятая
    views += int(divider[2]) #добавляем третий элимент
    if divider[1] not in users: #перебираем уникальных пользователей и добавляем в контейнер
        users.append(divider[1])
print(f'Среднее количество просмотров на уникального пользователя: {round(views/len(users),2)}')
```
