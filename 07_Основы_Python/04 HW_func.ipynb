```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}
```

```python
def number(number_doc):
    """
    Функция возвращает по номеру документа данные о его владельце
    """
    owner_name = ''
    for item in documents:
        if item['number'] == number_doc:
            owner_name = 'Владелец документа: ' + item['name']
    if owner_name == '': 
        owner_name = 'Документ не найден в базе'
    return owner_name
```

```python
def directory(number_doc):
    """
    Функция возвращает по номеру документа данные о месте хранения
    """
    shelf = ''
    for item, values in directories.items():
        if number_doc in values:
            shelf = f'Документ хранится на полке: {item}'
    if shelf == '':
        shelf='Документ не найден в базе'
    return shelf
```

```python
def list_docs():
    """
    Функция возвращает список всех документов на полках
    """
    for doc in documents:
        for item, values in directories.items():
            if doc['number'] in values:
                print('№ ' +doc['number'] + ',' + 'тип: '+ doc['type']+','+' владелец: '+ doc['name']+','+' полка хранения: '+ item)
```

```python
def list_of_shelfs():
    """
    Функция возвращает список полок
    """    
    list_shelf = ''
    for item, values in directories.items():
        list_shelf = list_shelf + item + ',' 
    list_shelf = list_shelf[:-1] + '.'
    return list_shelf
```

```python
def del_shelf(number_shelf):
    """
    Функция удаляе полку, если она не пустая
    """
    count = 0
    for item, values in directories.items():
        if number_shelf in item:
            count = 1
            if values != []:
                count = 2
    if count == 1:
        del directories[number_shelf]
        print(f'Полка удалена. Текущий перечень полок: {list_of_shelfs()}')
    elif count == 2:
        print(f'На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: {list_of_shelfs()}')
    else:
        print(f'Такая полка не существует. Текущий перечень полок: {list_of_shelfs()}')
```

```python
def add_shelf(number_shelf):
    """
    Функция добавляет полку, если такой ещё нет
    """    
    count = 0
    for item, values in directories.items():
        if number_shelf in item:
            count = 1
    if count == 1:
        print(f'Такая полка уже существует. Текущий перечень полок: {list_of_shelfs()}')
    else:
        directories[number_shelf]=[]
        print(f'Полка добавлена. Текущий перечень полок: {list_of_shelfs()}')
```

```python
def main():
    """
    Основная функция автоматизации работы секретаря
    """
    a = 0
    while a < 1:
        command_input=input('Введите команду: ')
        if command_input == 'p':
            number_input=input('Введите номер документа: ')
            print(number(number_input))
        elif command_input == 's':
            number_input=input('Введите номер документа: ')
            print(directory(number_input))
        elif command_input == 'l':
            list_docs()
        elif command_input == 'as':
            number_input=input('Введите номер полки: ')
            add_shelf(number_input)
        elif command_input == 'ds':
            number_input=input('Введите номер полки: ')
            del_shelf(number_input)
        elif command_input == 'q':
            break
        else:
            print('Нет такой команды')
```

```python
main()
```
