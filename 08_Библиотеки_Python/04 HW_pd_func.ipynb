# Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
df_ratings = pd.read_csv('/content/drive/MyDrive/ratings.csv')
```

```python
def ratings(row):
    if float(row['rating']) <= 2:
        return 'Low'
    if float(row['rating']) <= 4:
        return 'Middle'
    return 'Highe'
```

```python
df_movies = pd.read_csv('/content/drive/MyDrive/movies.csv')
```

```python
df_ratings = df_ratings.groupby('movieId').agg({'rating':'mean'}).reset_index()
df_ratings.head()
```

```python
df_ratings['level'] = df_ratings.apply(ratings, axis=1)
df_ratings = df_ratings.drop('rating', 1)
df_ratings.head()
```

```python
df_movies = df_movies.merge(df_ratings, on = 'movieId')
df_movies.head()
```

# Задание 2
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}

Результат классификации запишите в отдельный столбец region.

```python
df_keywords = pd.read_csv('/content/drive/MyDrive/keywords.csv')
df_keywords.head()
```

```python
geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
            'Северо-Запад': ['петербург', 'псков', 'мурманск'],
            'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск'] }
```

```python
def find_region(row):
    for i in geo_data.items():
        for a in i[1]:
            if a in row['keyword']:
                return  i[0]
    
    return 'undefined'
```

```python
df_keywords['region'] = df_keywords.apply(find_region, axis=1)
df_keywords.head()
```
