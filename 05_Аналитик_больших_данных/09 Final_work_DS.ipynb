Средствами Pandas, используя dataframe ratings, найдите id пользователя, поставившего больше всего оценок

```python
df_user = df_ratings[['user id', 'rating']].groupby('user id').sum()
```

Оставьте в датафрейме ratings только те фильмы, который оценил данный пользователь

```python
df_filtered = df_ratings.loc[df_ratings['user id'] == 450.0]
```

```python
df_filtered
```

Добавьте к датафрейму из задания 3 столбцы: \
a) По жанрам. Каждый столбец - это жанр. Единицу записываем, если фильм принадлежит данному жанру и 0 - если нет \
b) столбцы с общим количеством оценок от всех пользователей на фильм и суммарной оценкой от всех пользователей

```python
df_merged = pd.merge(df_filtered, df_movies, how='left', left_on='item id', right_on='movie id') 
```

```python
df_merged
```

```python
df_users_rating = df_ratings[['item id', 'rating']].groupby('item id', as_index=False).sum().rename(columns={'rating': 'ratings'})
```

```python
df_users_rating
```

```python
df_users_count = df_ratings[['item id', 'rating']].groupby('item id', as_index=False).count().rename(columns={'rating': 'count'})
```

```python
df_users_count
```

```python
df_users = pd.merge(df_users_rating, df_users_count, how = 'left', on = 'item id')
df_users
```

```python
df_best_user = pd.merge(df_merged, df_users, how='left', on='item id')
df_best_user
```

Для построения модели нам нужны признаки. В качестве таковых будем использовать:\
Год выхода \
Жанры \
Общее количество оценок \
Суммарную оценку


```python
df_best_user.columns
```

```python
X = df_best_user[['video release date', 'unknown', 'Action', 'Adventure', 'Animation', "Children's", 'Comedy ',
       'Crime ', 'Documentary ', 'Drama ', 'Fantasy ', 'Film-Noir', 'Horror ',
       'Musical ', 'Mystery ', 'Romance ', 'Sci-Fi', 'Thriller ', 'War ',
       'Western ', 'ratings', 'count']]
X
```

```python
y = df_best_user[['rating']]
y
```

Сформируйте X_train, X_test, y_train, y_test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

Создаем и обучаем модель линейной регресии

```python
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=False)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=False)
```

```python
lr.coef_
```

Получаем предсказание и оцениваем качество

```python
lr.predict(X_test)
```

Оцените качество, при помощи метода mean_squared_error для тестовой выборки

```python
mean_squared_error(y_test, lr.predict(X_test))
```

```python
mean_squared_error(y_train, lr.predict(X_train))
```

```python
#сортируем по популярным фильмам
df_popular = df_count_rat.sort(col('count').desc())
df_popular.show(5)
```

```python
#сортируем по непопулярным фильмам
df_unpopular = df_count_rat.sort(col('count').asc())
df_unpopular.show(5)
```
